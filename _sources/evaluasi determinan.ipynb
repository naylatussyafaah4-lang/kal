# Determinan Matriks dengan Ekspansi Baris


$$
\det(A)=\sum_{k=1}^{n}(-1)^{i+k}a_{ik}M_{ik}
$$

dengan:

$$
M_{ik}=\det(A_{ik})
$$

dan `Aik` adalah submatriks hasil menghapus baris ke-`i` dan kolom ke-`k`.

---

## 1)

$$
A=\begin{bmatrix}
-7 & -5\
1 & 4
\end{bmatrix}
$$

Pakai ekspansi baris ke-1:

$$
\det(A)=(-1)^{1+1}a_{11}M_{11}+(-1)^{1+2}a_{12}M_{12}
$$

Karena:

$$
a_{11}=-7,\quad a_{12}=-5
$$

maka:

$$
\det(A)=(-1)^2(-7)M_{11}+(-1)^3(-5)M_{12}
$$

### Cari minor

#### Minor `M11`

Hapus baris 1 kolom 1:

$$
A_{11}=[4]
$$

$$
M_{11}=\det(A_{11})=4
$$

#### Minor `M12`

Hapus baris 1 kolom 2:

$$
A_{12}=[1]
$$

$$
M_{12}=\det(A_{12})=1
$$

### Substitusi

$$
\det(A)=(-1)^2(-7)(4)+(-1)^3(-5)(1)
$$

$$
=1(-28)+(-1)(-5)
$$

$$
=-28+5=-23
$$

## Hasil nomor 1

$$
\boxed{\det(A)=-23}
$$

## 2)

$$
A=\begin{bmatrix}
0 & 2 & -3\
1 & -2 & -1\
0 & 0 & 1
\end{bmatrix}
$$

Pakai ekspansi baris ke-3 karena ada dua nol.

$$
\det(A)=\sum_{k=1}^{3}(-1)^{3+k}a_{3k}M_{3k}
$$

$$
\det(A)=(-1)^{3+1}a_{31}M_{31}+(-1)^{3+2}a_{32}M_{32}+(-1)^{3+3}a_{33}M_{33}
$$

Karena:

$$
a_{31}=0,\quad a_{32}=0,\quad a_{33}=1
$$

maka:

$$
\det(A)=(-1)^4(0)M_{31}+(-1)^5(0)M_{32}+(-1)^6(1)M_{33}
$$

$$
\det(A)=0+0+M_{33}
$$

### Cari minor `M33`
$$
$$
Hapus baris 3 kolom 3:

$$
A_{33}=
\begin{bmatrix}
0 & 2\
1 & -2
\end{bmatrix}
$$

$$
M_{33}=\det(A_{33})
===================

\begin{vmatrix}
0 & 2\
1 & -2
\end{vmatrix}
$$

$$
=(0)(-2)-(2)(1)
$$

$$
=0-2=-2
$$

Jadi:

$$
\det(A)=M_{33}=-2
$$

## Hasil nomor 2

$$
\boxed{\det(A)=-2}
$$

# Invers Matriks dengan Rumus Adjoin

Diketahui:

$$
(\operatorname{adj}A)*{ij}=(-1)^{i+j}M*{ji}
$$

$$
A^{-1}=\frac{1}{\det A}\operatorname{adj}A
$$

dengan (M_{ji}) adalah minor dari elemen pada baris (j), kolom (i).

---

## 4.

$$
A=\begin{bmatrix}
-7 & -5\
1 & 4
\end{bmatrix}
$$

### Determinan

$$
\det A=(-7)(4)-(-5)(1)
$$

$$
=-28+5=-23
$$

### Minor

$$
M_{11}=\det([4])=4
$$

$$
M_{12}=\det([1])=1
$$

$$
M_{21}=\det([-5])=-5
$$

$$
M_{22}=\det([-7])=-7
$$

### Adjoin

$$
(\operatorname{adj}A)*{11}=(-1)^{1+1}M*{11}=4
$$

$$
(\operatorname{adj}A)*{12}=(-1)^{1+2}M*{21}=-(-5)=5
$$

$$
(\operatorname{adj}A)*{21}=(-1)^{2+1}M*{12}=-(1)=-1
$$

$$
(\operatorname{adj}A)*{22}=(-1)^{2+2}M*{22}=-7
$$

Maka:

$$
\operatorname{adj}A=
\begin{bmatrix}
4 & 5\
-1 & -7
\end{bmatrix}
$$

### Invers

$$
A^{-1}=\frac{1}{-23}
\begin{bmatrix}
4 & 5\
-1 & -7
\end{bmatrix}
$$

$$
A^{-1}=
\begin{bmatrix}
-\frac{4}{23} & -\frac{5}{23}\
\frac{1}{23} & \frac{7}{23}
\end{bmatrix}
$$

### Hasil nomor 4

$$
\boxed{
A^{-1}=
\begin{bmatrix}
-\frac{4}{23} & -\frac{5}{23}\
\frac{1}{23} & \frac{7}{23}
\end{bmatrix}}
$$

## 5.

$$
A=
\begin{bmatrix}
0 & 2 & -3\
1 & -2 & -1\
0 & 0 & 1
\end{bmatrix}
$$

### Determinan

Ekspansi baris ke-3:

$$
\det A=
(-1)^{3+1}(0)M_{31}
+
(-1)^{3+2}(0)M_{32}
+
(-1)^{3+3}(1)M_{33}
$$

$$
\det A=M_{33}
$$

$$
M_{33}=
\begin{vmatrix}
0 & 2\
1 & -2
\end{vmatrix}
=(0)(-2)-(2)(1)=-2
$$

Jadi:

$$
\det A=-2
$$

### Minor

$$
M_{11}=
\begin{vmatrix}
-2 & -1\
0 & 1
\end{vmatrix}
=-2
$$

$$
M_{12}=
\begin{vmatrix}
1 & -1\
0 & 1
\end{vmatrix}
=1
$$

$$
M_{13}=
\begin{vmatrix}
1 & -2\
0 & 0
\end{vmatrix}
=0
$$

$$
M_{21}=
\begin{vmatrix}
2 & -3\
0 & 1
\end{vmatrix}
=2
$$

$$
M_{22}=
\begin{vmatrix}
0 & -3\
0 & 1
\end{vmatrix}
=0
$$

$$
M_{23}=
\begin{vmatrix}
0 & 2\
0 & 0
\end{vmatrix}
=0
$$

$$
M_{31}=
\begin{vmatrix}
2 & -3\
-2 & -1
\end{vmatrix}
=(2)(-1)-(-3)(-2)=-2-6=-8
$$

$$
M_{32}=
\begin{vmatrix}
0 & -3\
1 & -1
\end{vmatrix}
=(0)(-1)-(-3)(1)=3
$$

$$
M_{33}=
\begin{vmatrix}
0 & 2\
1 & -2
\end{vmatrix}
=-2
$$

### Adjoin

$$
(\operatorname{adj}A)*{11}=(-1)^{1+1}M*{11}=-2
$$

$$
(\operatorname{adj}A)*{12}=(-1)^{1+2}M*{21}=-(2)=-2
$$

$$
(\operatorname{adj}A)*{13}=(-1)^{1+3}M*{31}=-8
$$

$$
(\operatorname{adj}A)*{21}=(-1)^{2+1}M*{12}=-(1)=-1
$$

$$
(\operatorname{adj}A)*{22}=(-1)^{2+2}M*{22}=0
$$

$$
(\operatorname{adj}A)*{23}=(-1)^{2+3}M*{32}=-(3)=-3
$$

$$
(\operatorname{adj}A)*{31}=(-1)^{3+1}M*{13}=0
$$

$$
(\operatorname{adj}A)*{32}=(-1)^{3+2}M*{23}=0
$$

$$
(\operatorname{adj}A)*{33}=(-1)^{3+3}M*{33}=-2
$$

Maka:

$$
\operatorname{adj}A=
\begin{bmatrix}
-2 & -2 & -8\
-1 & 0 & -3\
0 & 0 & -2
\end{bmatrix}
$$

### Invers

$$
A^{-1}=\frac{1}{-2}
\begin{bmatrix}
-2 & -2 & -8\
-1 & 0 & -3\
0 & 0 & -2
\end{bmatrix}
$$

$$
A^{-1}=
\begin{bmatrix}
1 & 1 & 4\
\frac{1}{2} & 0 & \frac{3}{2}\
0 & 0 & 1
\end{bmatrix}
$$

### Hasil nomor 5

$$
\boxed{
A^{-1}=
\begin{bmatrix}
1 & 1 & 4\
\frac{1}{2} & 0 & \frac{3}{2}\

## 6.

$$
A=
\begin{bmatrix}
1 & -3 & 1 & 1\
-3 & 1 & 1 & 1\
1 & 1 & -3 & 1\
1 & 1 & 1 & -3
\end{bmatrix}
$$

### Cek determinan

Jumlah setiap baris:

$$
1-3+1+1=0
$$

$$
-3+1+1+1=0
$$

$$
1+1-3+1=0
$$

$$
1+1+1-3=0
$$

Maka:

$$
A
\begin{bmatrix}
1\
1\
1\
1
\end{bmatrix}
=============

\begin{bmatrix}
0\
0\
0\
0
\end{bmatrix}
$$

Karena ada vektor tak nol yang menghasilkan vektor nol, maka matriks singular.

$$
\det A=0
$$

### Kesimpulan

$$
A^{-1}=\frac{1}{\det A}\operatorname{adj}A
$$

Karena (\det A=0), maka invers tidak ada.

### Hasil nomor 6

$$
\boxed{A^{-1}\text{ tidak ada}}
$$
