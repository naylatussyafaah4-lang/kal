#  Contoh 1: SPL Tiga Variabel (3x3)

## Soal

```
x + y + z = 6
2x + y + z = 9
x + 2y + 3z = 14
```

---

##  Langkah 1: Bentuk Matriks Augmented

```
| 1  1  1 |  6 |
| 2  1  1 |  9 |
| 1  2  3 | 14 |
```

---

##  Langkah 2: Eliminasi Kolom Pertama

Operasi baris:

```
R2 → R2 - 2R1
R3 → R3 - R1
```

Hasil:

```
| 1  1  1 |  6 |
| 0 -1 -1 | -3 |
| 0  1  2 |  8 |
```

---

##  Langkah 3: Eliminasi Kolom Kedua

Operasi baris:

```
R3 → R3 + R2
```

Hasil:

```
| 1  1  1 |  6 |
| 0 -1 -1 | -3 |
| 0  0  1 |  5 |
```

---

##  Langkah 4: Substitusi Balik

Dari baris ketiga:

```
z = 5
```

Substitusi ke baris kedua:

```
- y - z = -3
- y - 5 = -3
- y = 2
y = -2
```

Substitusi ke baris pertama:

```
x + y + z = 6
x - 2 + 5 = 6
x + 3 = 6
x = 3
```

---

##  Hasil Akhir

```
x = 3
y = -2
z = 5
```

#  Contoh 2: SPL Empat Variabel (4x4)

## Soal

```
x + y + z + w = 10
2x + y + z + w = 13
x + 2y + 3z + w = 18
x + y + z + 2w = 13
```

---

##  Langkah 1: Matriks Augmented

```
| 1  1  1  1 | 10 |
| 2  1  1  1 | 13 |
| 1  2  3  1 | 18 |
| 1  1  1  2 | 13 |
```

---

##  Langkah 2: Eliminasi Kolom Pertama

Operasi baris:

```
R2 → R2 - 2R1
R3 → R3 - R1
R4 → R4 - R1
```

Hasil:

```
| 1  1  1  1 | 10 |
| 0 -1 -1 -1 | -7 |
| 0  1  2  0 |  8 |
| 0  0  0  1 |  3 |
```

---

##  Langkah 3: Eliminasi Kolom Kedua

Operasi baris:

```
R3 → R3 + R2
```

Hasil:

```
| 1  1  1  1 | 10 |
| 0 -1 -1 -1 | -7 |
| 0  0  1 -1 |  1 |
| 0  0  0  1 |  3 |
```

---

##  Langkah 4: Substitusi Balik

Dari baris keempat:

```
w = 3
```

Baris ketiga:

```
z - w = 1
z - 3 = 1
z = 4
```

Baris kedua:

```
- y - z - w = -7
- y - 4 - 3 = -7
- y - 7 = -7
y = 0
```

Baris pertama:

```
x + y + z + w = 10
x + 0 + 4 + 3 = 10
x + 7 = 
```
